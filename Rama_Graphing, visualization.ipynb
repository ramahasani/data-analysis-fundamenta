iterate over the region to make the 25 moving average as the vallue of the region

```python
graph_df = pd.DataFrame()

for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df["region"]==region]
    region_df.set_index("Date", inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}_price25ma"] = region_df["AveragePrice"].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f"{region}_price25ma"]]#2brackets to be a df
        
    else:
        graph_df = graph_df.join(region_df[f"{region}_price25ma"])
```

do the same thing as above, but now with the organic type and remove the 16 limit so it conclude all important data

```python
import pandas as pd

df = pd.read_csv("D:/Rama/AI Coaching/avocado.csv")
df = df.copy()[df['type']=='organic']
df["Date"] = pd.to_datetime(df["Date"])
df.sort_values(by="Date", ascending=True, inplace=True)

graph_df = pd.DataFrame()

for region in df['region'].unique():
    print(region)
    region_df = df.copy()[df['region']==region]
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f"{region}_price25ma"] = region_df["AveragePrice"].rolling(25).mean()

    if graph_df.empty:
        graph_df = region_df[[f"{region}_price25ma"]] 
    else:
        graph_df = graph_df.join(region_df[f"{region}_price25ma"])

graph_df.tail()
```

plot the graph df

```python
graph_df.dropna().plot(figsize=(8,5), legend=False)
```
